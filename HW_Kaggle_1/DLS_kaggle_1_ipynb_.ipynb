## Применение линейных моделей (3 балла)

1) Обработайте данные для того, чтобы к ним можно было применить LogisticRegression. Т.е. отнормируйте числовые признаки, а категориальные закодируйте с помощью one-hot-encoding'а. 

2) С помощью кроссвалидации или разделения на train/valid выборку протестируйте разные значения гиперпараметра C и выберите лучший (можно тестировать С=100, 10, 1, 0.1, 0.01, 0.001) по метрике ROC-AUC. 

Если вы разделяете на train/valid, то используйте LogisticRegressionCV. Он сам при вызове .fit() подберет параметр С. (не забудьте передать scroing='roc_auc', чтобы при кроссвалидации сравнивались значения этой метрики, и refit=True, чтобы при потом модель обучилась на всем датасете с лучшим параметром C). 


(более сложный вариант) Если вы будете использовать кроссвалидацию, то преобразования данных и LogisticRegression нужно соединить в один Pipeline с помощью make_pipeline, как это делалось во втором семинаре. Потом pipeline надо передать в GridSearchCV. Для one-hot-encoding'a можно испльзовать комбинацию LabelEncoder + OneHotEncoder (сначала превращаем строчки в числа, а потом числа првращаем в one-hot вектора.)

```python
def search_and_draw(X, y, model):
  X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.5, random_state=42)
  
  y_train_predicted = model.predict_proba(X_train)[:, 1]
  y_test_predicted = model.predict_proba(X_test)[:, 1]

  train_roc_auc = roc_auc_score(y_train, y_train_predicted)
  test_roc_auc = roc_auc_score(y_test, y_test_predicted)

  plt.figure(figsize=(20,10))
  plt.plot(*roc_curve(y_train, y_train_predicted)[:2], label='train AUC={:.4f}'.format(train_roc_auc))
  plt.plot(*roc_curve(y_test, y_test_predicted)[:2], label='test AUC={:.4f}'.format(test_roc_auc))
  legend_box = plt.legend(fontsize='large', framealpha=1).get_frame()
  legend_box.set_facecolor("white")
  legend_box.set_edgecolor("black")
  plt.plot(np.linspace(0,1,100), np.linspace(0,1,100))
  plt.show()
```

```python
# Для ускорения выполнения блокнота были подставлены лучшие гиперпараметры, найденные с помощью GridSearch.
numeric_transformer = Pipeline(steps=[
    ("scaler", StandardScaler())
])
preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, num_cols),
        ("cat", OneHotEncoder(handle_unknown='ignore'), cat_cols)
    ]
)
params = {
    'model__penalty': ['l1'],
    'model__C': [0.1],
    'model__solver': ['saga']
}
clf = Pipeline(steps=[
    ("prep", preprocessor),
    ("model", LogisticRegression(max_iter=1000, random_state=42)),
])
grid_search = GridSearchCV(
    estimator=clf,
    param_grid=params,
    scoring='roc_auc',
    n_jobs=-1,
    cv=5,
    refit=True
)
logreg = grid_search.fit(X_train, y_train)
```

```python
search_and_draw(X_train, y_train, logreg)
```

```python
print(f'best_params: {logreg.best_params_}')
print(f'best_score: {logreg.best_score_}')
```

```python
numeric_transformer = Pipeline(steps=[
    ("scaler", StandardScaler())
])
preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, num_cols),
        ("cat", OneHotEncoder(handle_unknown='ignore'), cat_cols)
    ]
)
clf = Pipeline(steps=[
    ("prep", preprocessor),
    ("model", RandomForestClassifier()),
])
rf_params = dict(     
    model__max_depth = [11], #range(9, 16),
    model__min_samples_split = [7], # range(4, 13),
    model__min_samples_leaf = [15], # range(3, 20),   
    model__n_estimators = [60] #range(10, 70, 10)
)
grid_search = GridSearchCV(
    estimator=clf,
    param_grid=rf_params,
    scoring="roc_auc",
    n_jobs=-1,
    cv=10,
    refit=True,
)
rf = grid_search.fit(X_train, y_train)
```

```python
meta = CatBoostClassifier(
    logging_level='Silent',
    eval_metric="AUC:hints=skip_train~false",
    metric_period=1000,
    random_seed=42,
    grow_policy="Depthwise",
    l2_leaf_reg=100,
    learning_rate=0.05,
    max_depth=6,
    min_data_in_leaf=10,
    n_estimators=1000,
    random_strength=5,
    subsample=0.6,
)
stacking = StackingClassifier(
    estimators=[
        ('logreg', logreg),
        ("rf", rf),
        ("catboost", catboost)
        
    ],
    final_estimator=meta,
    n_jobs=-1,
)

stacking.fit(X_train, y_train)
```

```python
y_pred = stacking.predict_proba(X_test)[:, 1]
search_and_draw(X_train, y_train, stacking)
submission = pd.read_csv("./submission.csv")
submission['Churn'] = y_pred
submission.to_csv("./Ssubmission.csv", index=False)
```
